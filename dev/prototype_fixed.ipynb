# Fixed Quantum Circuit Prototype

Quick demo: 2 qubits, H on q0 (superposition), measure shots. Fixed __file__ error for Jupyter/VSCode.

```python
import torch
import sys
import os
# Fixed path for Jupyter/VSCode (relative to project root)
project_root = '/home/hari/Development/github/personal/QuantumSimulation'
sys.path.insert(0, project_root)

from src import BaseQuantumCircuit, Zero, H

torch.manual_seed(42)
```

```python
# 2-qubit circuit: H on q0
circuit = BaseQuantumCircuit()
circuit.add_qubit(Zero())
circuit.add_qubit(Zero())
circuit.add_gate(H(), target=0)

state = circuit.forward()
probs = torch.real(state * state.conj()).squeeze()
print('State shape:', state.shape)
print('Probs:', probs)
print('Expected: ~[0.25]*4')
```

```python
# Measure 100 shots
outcomes = [circuit.measure(num_shots=1)[0] for _ in range(100)]
from collections import Counter
print('Measured counts:', Counter(outcomes))
```

```python
# H on q1 too
circuit.add_gate(H(), target=1)
probs2 = torch.real(circuit.forward() * circuit.forward().conj()).squeeze()
print('After H q1:', probs2)
```
