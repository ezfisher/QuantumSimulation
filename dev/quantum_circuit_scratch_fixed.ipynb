```python
import torch
import sys
import os
# Absolute path to project root (works in Jupyter/VSCode)
project_root = '/home/hari/Development/github/personal/QuantumSimulation'
sys.path.insert(0, project_root)

from src import BaseQuantumCircuit, Zero, H

torch.manual_seed(42)
```

```python
circuit = BaseQuantumCircuit()
circuit.add_qubit(Zero())
circuit.add_qubit(Zero())
circuit.add_gate(H(), 0)

state = circuit.forward()
probs = torch.real(state * torch.conj(state)).squeeze()
print('Shape:', state.shape)
print('Probs H q0:', probs)
print('Expected approx [0.25, 0.25, 0.25, 0.25]')
```

```python
outcomes = [circuit.measure(num_shots=1)[0] for _ in range(100)]
from collections import Counter
print('100 shots:', Counter(outcomes))
```

```python
circuit.add_gate(H(), 1)
probs2 = torch.real(circuit.forward() * torch.conj(circuit.forward())).squeeze()
print('After H q1:', probs2)
```
