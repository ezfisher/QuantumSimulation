# Quantum Circuit Prototype

Quick demo using BaseQuantumCircuit: 2 qubits, H on q0 → separable superposition H|0⟩ ⊗ |0⟩.

```python
import torch
import sys
import os
sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath(__file__))))

from src import BaseQuantumCircuit, Zero, H

torch.manual_seed(42)
```

```python
# Step 1: 2-qubit circuit, H on q0
circuit = BaseQuantumCircuit()
circuit.add_qubit(Zero())
circuit.add_qubit(Zero())
circuit.add_gate(H(), target=0)

state = circuit.forward()
probs = torch.real(state * state.conj()).squeeze()
print('State shape:', state.shape)
print('Probs:', probs)
print('Expected: ~[0.25, 0.25, 0.25, 0.25] (H|00> = (|0>+|1>)|0>/2)')
```

```python
# Measure (multi-shot for stats)
outcomes = []
for _ in range(100):
    outcome, _, _ = circuit.measure(num_shots=1)
    outcomes.append(outcome)

unique, counts = torch.unique(torch.tensor(outcomes), return_counts=True)
print('Measured:', dict(zip(unique.tolist(), counts.tolist())))
```

```python
# Step 2: Add H on q1 → full uniform
circuit.add_gate(H(), target=1)
state2 = circuit.forward()
probs2 = torch.real(state2 * state2.conj()).squeeze()
print('Probs after H q1:', probs2)
```
